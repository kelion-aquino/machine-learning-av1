```python
df_embarked = df.groupby('Embarked')

df_embarked.mean()
```

```python
# Realizando o Agrupamento dos dados pela coluna 'Embarked' e calculando a média sobreviventes
df.groupby('Embarked')['Survived'].mean()
```

```python
# Realizando o agrupamento dos dados pela coluna 'Age' e calculando o maior valor pago pela passagem
df.groupby('Age')['Fare'].max()
```

```python
# Selecionar linhas que tenham pelo menos um valor ausente ou nulo
df[df.isnull().any(axis=1)].head(100)
```

```python
# Filtar todas as linhas onde o valor 'Cabin' estão faltando ou nulos:
df1 = df[df['Cabin'].notnull( )]
df1.head(100)
```

```python
# Remover todas as observações com valores nulos
df2 = df1.dropna()

df2.head(10)
```

```python
# Preenchendo os valores ausentes com zeros
nomiss = df['Age'].fillna(df['Age'].mean())
nomiss.isnull().any()
```

```python
# Mostrando Gráficos com Python no Notebook
%matplotlib inline
```

```python
# Usando Matplotlib para desenhar um histogram para a coluna 'Survived'
plt.hist(df['Survived'],bins=8)
```

```python
# Usando Matplotlib para desenhar um histogram para a coluna 'Age'
sns.distplot(df['Age'])
```

```python
# Uso do matplotlib para mostrar o "Gráfico de Barras" barplot
df.groupby(['Sex'])['Survived'].mean().plot(kind='pie')
```

```python
# Uso do seaborn para mostrar o Gráfico de Barras
sns.set_style("whitegrid")
ax = sns.barplot(x='Embarked', y='Survived', data=df, estimator=len)
```

```python
# Dividindo o gráfico em 02 grupos:
ax = sns.barplot(x='Embarked', y='Age', hue='Survived', data=df, estimator=len)
```

```python
#Violinplot
sns.violinplot(x="Age", data=df)
```

```python
#Scatterplot in seaborn
sns.jointplot(x='Age', y='Fare', data=df)
```

```python
# Representar a regressão linear para 2 variáveis numéricas - regplot
sns.regplot(x='Age', y='Fare', data=df)
```

```python
# Detectar Out-lier
sns.boxplot(x='Survived', y='Age', hue='Sex', data=df)
```

```python
# Side-by-side box plot
sns.boxplot(x='Embarked', y='Pclass', data=df, hue='Sex')
```

```python
# Pairplot 
sns.pairplot(df)
```
